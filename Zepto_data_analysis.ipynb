```python
print("What is the maximum and minimum MRP of products in the dataset?")
print(f"Maximum MRP: {df['mrp'].max()}")
print(f"Minimum MRP: {df['mrp'].min()}")
```

```python
print(f"How many products are out of stock?")
df['outOfStock'].sum()
```

```python
print("Which products have the highest discount percentage?")
max_discount = df['discountPercent'].max()
highest_discount_products = df[df['discountPercent'] == max_discount]
highest_discount_products[['name', 'discountPercent','Category']]
```

```python
print("Which products have the lowest discount percentage?")
min_discount = df['discountPercent'].min()
lowest_discount_products = df[df['discountPercent'] == min_discount]
lowest_discount_products[['name', 'discountPercent','Category']]
```

```python
print("Find the top 10 products with the highest discounted selling price.")
top_10_discounted_highest = df.nlargest(10,'discountedSellingPrice')
top_10_discounted_highest[['name', 'discountedSellingPrice','Category']]
```

```python
print("Find the top 10 products with the lowest discounted selling price.")
top_10_lowest_discounted = df.nsmallest(10,'discountedSellingPrice')
top_10_lowest_discounted[['name', 'discountedSellingPrice','Category']]
```

```python
print("Which product categories have the most items available in stock?")
df.groupby('Category')['quantity'].max()
```

```python
print("What is the total available quantity of products in each category?")
df.groupby("Category")['availableQuantity'].sum()
```

```python
print("Which product has the maximum available quantity?")
max_available_product = df['availableQuantity'].max()
df['Category'][max_available_product]
```

```python
print("Which product has the minimum available quantity?")
min_available_product = df['availableQuantity'].min()
df['Category'][min_available_product]
```

```python
print("What is the average weight of products across categories?")
avg_weight_by_category = df.groupby('Category')['weightInGms'].mean()
print(avg_weight_by_category)
```

```python
print('Which category has the heaviest product?')
heaviest_product = df.loc[df['weightInGms'].idxmax()]
heaviest_category = heaviest_product['Category']
heaviest_weight = heaviest_product['weightInGms']
print(f'The heaviest product with the heaviest category is {heaviest_category} with weight of {heaviest_weight} grams ')
```

```python
print('Which category has the lightest product?')
lightest_product  = df.loc[df['weightInGms'].idxmin()]
lightest_category = lightest_product['Category']
lightest_weight = lightest_product['weightInGms']
print(f'The lightest product in the category is {lightest_category} with {lightest_weight} grams')
```

```python
print('Find the products where the discounted selling price is greater than the MRP (data error check).')
error_products = (df['discountedSellingPrice'] > df['mrp']).sum()
print(error_products)
```

```python
print('How many products have no discount (discountPercent = 0)?')
product_having_0_dis = (df['discountPercent'] == 0).sum()
print(f"{product_having_0_dis} number products having 0 discount percentage.")
```

```python
print('What is the category-wise average discount percentage?')
Avg_dis_per = df.groupby('Category')['discountPercent'].mean()
print(Avg_dis_per)
```

```python
print('Find the distribution of products based on their weight range (e.g., <250g, 250-500g, 500-1000g, >1000g).')

bins = [0,250,500,1000,df['weightInGms'].max()]
labels = ['<250g', '250-500g','500-1000g','>1000g']

df['weight_range'] = pd.cut(df['weightInGms'],bins = bins , labels = labels , include_lowest = True)

distribution = df['weight_range'].value_counts().sort_index()

print(distribution)
```

```python
plt.figure(figsize=(7,4))
distribution.plot(kind='bar', color='skyblue', width=0.6)
plt.xlabel('Weight Range')
plt.ylabel('Number of Products')
plt.title('Product Weight Range Distribution')
plt.tight_layout()
plt.show()
```

```python
print('How many products are priced above â‚¹500?')
Product_abv_500 = (df['mrp']>500).sum()
print(Product_abv_500," Products")
```

```python
print('How many products are priced below â‚¹100?')
Price_below_100 = (df['mrp']<100).sum()
print(Price_below_100," Products")
```

```python
print('Which category contributes the highest revenue potential (MRP Ã— availableQuantity)?')

df['revenue_potential'] = df['mrp'] * df['availableQuantity']

category_revenue = df.groupby('Category')['revenue_potential'].sum()

top_category = category_revenue.idxmax()
top_revenue = category_revenue.max()

# Print the result
print(f"Category with highest revenue potential: {top_category} ({top_revenue})")
```

```python
print('Which category contributes the lowest revenue potential?')

low_category = category_revenue.idxmin()
low_revenue = category_revenue.min()
print(f"Category with the lowest revenue potential : {low_category} ({low_revenue})")
```

```python
print('List the top 5 categories with the highest average selling price after discount.')

highest_avg_sell_price = df.groupby('Category')['discountedSellingPrice'].mean()

high_avg_sell_category = highest_avg_sell_price.sort_values(ascending=False)
Top_5_high_avg_sell_category = high_avg_sell_category.head(5)
print(Top_5_high_avg_sell_category)
```

```python
Top_5_high_avg_sell_category.plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Average Discounted Selling Price')
plt.title('Top 5 Categories by Average Discounted Selling Price')
plt.show()
```

```python
print('List the top 5 categories with the lowest average selling price after discount.')

top_5_low_avg_sell_price = df.groupby('Category')['discountedSellingPrice'].mean().sort_values().head(5)
print(top_5_low_avg_sell_price)
```

```python
top_5_low_avg_sell_price.plot(kind='bar',color = 'skyblue')
plt.xlabel('Category')
plt.ylabel('Average Discounted Selling Price')
plt.title('Top 5 categories with the lowest average selling price after discount')
plt.show()
```

```python
print('What is the total revenue loss due to discounts (sum of (MRP - discountedSellingPrice) Ã— quantity)?')

Total_revenue_loss = ((df['mrp']-df['discountedSellingPrice'])*df['availableQuantity']).sum()
print(f'Total revenue loss due to discount : {Total_revenue_loss}')
```
